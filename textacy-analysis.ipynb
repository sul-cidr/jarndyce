```python
import textacy
import spacy
import os
```

```python
!python -m spacy download en
nlp = spacy.load('en')
```

```python
def text_stream(list_of_texts):
    for fn in list_of_texts:
        with open(fn, 'r') as f:
            yield f.read()    
```

```python
corpus_dir = "tm_texts"
corpus_dir_path = os.fsencode(corpus_dir)

filenames = []
for file in os.listdir(corpus_dir_path):
    filename = os.fsdecode(file)
    if filename.endswith(".txt"): 
        filenames.append("{}/{}".format(corpus_dir, filename))
    else:
        continue
```
