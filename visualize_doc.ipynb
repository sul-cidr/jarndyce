```python
def plot_doc_topics(df, book_id):
    single = df[df["source_id"] == book_id]
    single = single.drop(["source_id"], axis=1)
    title = single["Title"][0]
    single = single.drop(["Title"], axis=1)
    trans = single.T
    trans = trans[trans[0] > 0.00001]
    trans = trans.rename(columns={0: "proportion"})
    trans["topic"] = trans.index
    trans.index.name = "topic"
    fig = sns.barplot(x="topic", y="proportion",  data=trans)
    fig.set_title(title)
    return fig
```

```python
from sklearn.manifold import TSNE
```

```python
df
```

```python
lda_data = df.drop(["source_id", "Title"], axis=1).as_matrix()
```

```python
# tsne + bokeh based on https://www.kaggle.com/yohanb/lda-visualized-using-t-sne-and-bokeh#
```

```python
tsne = TSNE(perplexity=15)
tsne_embedding = tsne.fit_transform(lda_data)
```

```python
tsne_df = pd.DataFrame(tsne_embedding, columns=["x", "y"])
```

```python
tsne_df['hue'] = lda_data.argmax(axis=1)
```

```python
tsne_df
```

```python
from bokeh.palettes import viridis
```

```python
source = ColumnDataSource(
    data=dict(
        x = tsne_df.x,
        y = tsne_df.y,
        titles = df["Title"],
        source_id = df["source_id"],
        colors = [viridis(60)[i] for i in tsne_df.hue],
        hue = tsne_df.hue
    ))
```

```python
tooltips = [
    ('titles', '@titles'),
    ('source_id', '@source_id'),
    ('topic', '@hue')
]
```

```python
plot_title = "t-SNE embedding of topic model"
```

```python
plot_tsne = figure(plot_width=800, plot_height=800, title=plot_title, tooltips=tooltips)
```

```python
plot_tsne.scatter("x", "y", source=source, fill_color="colors", fill_alpha=0.8, size=10)
```

```python
show(plot_tsne)
```
