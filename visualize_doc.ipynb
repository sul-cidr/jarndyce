```python
def plot_doc_topics(df, book_id):
    single = df[df["source_id"] == book_id]
    single = single.drop(["source_id"], axis=1)
    title = single["Title"][0]
    single = single.drop(["Title"], axis=1)
    trans = single.T
    trans = trans[trans[0] > 0.00001]
    trans = trans.rename(columns={0: "proportion"})
    trans["topic"] = trans.index
    trans.index.name = "topic"
    fig = sns.barplot(x="topic", y="proportion",  data=trans)
    fig.set_title(title)
    return fig
```

```python
plot_doc_topics(df, 36105213320000)
```

```python
def make_doc_iplot(df, book_id):
    single = df[df["source_id"] == book_id]
    single = single.drop(["source_id"], axis=1)
    title = single["Title"][0]
    single = single.drop(["Title"], axis=1)
    trans = single.T
    trans = trans[trans[0] > 0.00001]
    trans = trans.rename(columns={0: "proportion"})
    trans["topic"] = trans.index
    trans.index.name = "topic"
    x_vals = trans["topic"].tolist()
    p = figure(x_range = x_vals, plot_height=500, plot_width=800, title=title)
    p.vbar(x=x_vals, top=trans["proportion"].tolist(), width=0.9)
    show(p)
    return p
```

```python
make_doc_iplot(df, 36105213320000)
```
