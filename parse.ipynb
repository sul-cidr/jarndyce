```python
from bs4 import BeautifulSoup
import glob
import pandas as pd
import zipfile
import os
```

```python
from tqdm import tqdm
```

```python
# unzip all of the zipfiles and put them flatly in an xml dir
for fn in glob.glob('batch*/*.zip'):
    xml_dir = 'xml'
    zipper = zipfile.ZipFile(fn, 'r')
    zipper.extractall(xml_dir)
    print(fn)
```

```python
# function to parse the xml files and pull out text content
def parse_xml(fn):
    with open(fn, 'r') as f:
        soup = BeautifulSoup(f)
        els = soup.find_all(content=True)
        text = " ".join([elem["content"] for elem in els])
    return text
```

```python
parse_xml('xml/36105213320000_0016.xml')
```

```python
# Given a filename, extract just the book id
def get_book_id(fn):
    return os.path.split(fn)[1].split('_')[0]
```

```python
# Put together a set of book ids
book_ids = {get_book_id(fn) for fn in glob.glob(os.path.join('xml', '*.xml'))}
len(book_ids)
```

```python
# Experiment with one book id to make sure it works
with open('plain/36105213321313.txt', 'w') as f:
    for fn in sorted(glob.glob(os.path.join('xml', '36105213321313_*.xml'))):
        print(fn)
        f.write(parse_xml(fn))
```

```python
import multiprocessing as mp
```

```python
# function to take a book id, find all xml files related, process, and write to txt
def write_book(book_id):
    with open(os.path.join('plain', book_id + '.txt'), 'w') as f:
        for fn in sorted(glob.glob(os.path.join('xml', book_id + '_*.xml'))):
            f.write(parse_xml(fn))
    return book_id
```

```python
# Since this would take a while for all 1417 books, let's create a pool to handle the work
pool = mp.Pool(8)
```

```python
pool.map(write_book, book_ids)
```

```python
pool.close()
```
