```python
import pandas as pd
from bokeh.io import output_notebook, show, save
output_notebook()
```

```python
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
```

```python
df = pd.read_csv('vis-data.csv')
df = df.drop('Unnamed: 0', axis=1)
```

```python
df
```

```python
words = pd.read_csv('full-60-topic-words.txt')
```

```python
def make_doc_iplot(df, words_df, idx, book_id):
    single = df[df["source_id"] == book_id]
    single = single.drop(["source_id"], axis=1)
    title = single["Title"].tolist()[0]
    single = single.drop(["Title"], axis=1)
    trans = single.T
    trans = trans[trans[idx] > 0.0001]
    trans = trans.rename(columns={idx: "proportion"})
    trans["topic"] = trans.index
    trans.index.name = "topic"
    x_vals = trans["topic"].tolist()
    result = pd.merge(trans, words, left_on="topic", right_on="topic")
    source = ColumnDataSource(result)
    tooltips = [
        ("proportion", "@proportion"),
        ("words", "@words"),
    ]
    p = figure(x_range = x_vals, plot_height=500, plot_width=1000, title=title, tooltips=tooltips)
    p.vbar("topic", top="proportion", width=0.9, source=source)
#     show(p)
    return p
```

```python
plot = make_doc_iplot(df, words, 0, 36105213320000)
```

```python
novels = df["source_id"].tolist()
```

```python
for idx, novel in enumerate(novels):
    p = make_doc_iplot(df, words, idx, novel)
    save(p, "novel_images/{}.html".format(novel))
```
